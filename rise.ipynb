```python
import requests
import re
import numpy as np
from bs4 import BeautifulSoup

# p for slice length
def gen_slices(Lines,p):
    lines=[]
    for l in Lines:
        lines.append(l.strip(" "))
    slices=[]
    j=0
    n=len(lines)
    print("No of lines:",n)
    k=int(n/p)
    for i in range(0,k):
        slices.append(lines[(i*p):(i*p)+p])
    slices.append(lines[k*p:])
    return slices   

#Getting Page From Internet
url="https://www.hackerearth.com/practice/python/getting-started/input-and-output/tutorial/"
home="https://www.hackerearth.com"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
data=str(soup)
slice_len=5
slices=gen_slices(data.splitlines(),slice_len)
print("No of sllices:",len(slices))
```
